```python
clean_data.loc[clean_data['Køkkenforhold'].isna(),
             'Køkkenforhold'] = clean_data.groupby(["address.postalId", 'numberOfRooms'])[
    'Køkkenforhold'].transform(lambda x: x.mode()[0] if any(x.mode()) else np.nan)
clean_data.loc[clean_data['Ydervægsmateriale'].isna(),
             'Ydervægsmateriale'] = clean_data.groupby(["address.postalId", 'numberOfRooms'])[
    'Ydervægsmateriale'].transform(lambda x: x.mode()[0] if any(x.mode()) else np.nan)
clean_data.loc[clean_data['Tagmateriale'].isna(),
             'Tagmateriale'] = clean_data.groupby(["address.postalId", 'numberOfRooms'])[
    'Tagmateriale'].transform(lambda x: x.mode()[0] if any(x.mode()) else np.nan)
clean_data.loc[clean_data['Varmeinstallation'].isna(),
             'Varmeinstallation'] = clean_data.groupby(["address.postalId", 'numberOfRooms'])[
    'Varmeinstallation'].transform(lambda x: x.mode()[0] if any(x.mode()) else np.nan)
```

### Economic variables

Statbank house price index
From statbank EJ14 is downloaded, to get an index for house/apartment prices for every address

```python
price_index = pd.read_csv('salesPriceIndex.csv', delimiter=',', header=1, skiprows=1)
price_index = price_index.drop([0, 1]).T

houseIndex = {index.replace('M', '-'): value for index, value in price_index[2].iteritems()}
apartmentIndex = {index.replace('M', '-'): value for index, value in price_index[3].iteritems()}
```

```python
clean_data = clean_data[~(clean_data['address.latestSale.salesYear'] == 2022)]
```

```python
clean_data['previousMonth'] = clean_data['address.latestSale.saleDate'] - pd.DateOffset(months=1)
clean_data['previousMonth'] = clean_data['previousMonth'].astype(str).str[:7]
clean_data['priceHouse'] = clean_data['previousMonth'].replace(houseIndex)
clean_data['priceApartment'] = clean_data['previousMonth'].replace(apartmentIndex)

def f(X):
    if X['address.itemTypeName'] == 'Ejerlejlighed':
        priceIndex = X['priceApartment']
    else:
        priceIndex = X['priceHouse']
    return priceIndex
clean_data['priceIndex'] = clean_data.apply(f, axis=1)
clean_data = clean_data.drop(columns=['priceApartment', 'priceHouse'])
```
